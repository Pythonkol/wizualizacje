Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystykami

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)

# Statystyki opisowe
print("Statystyki dla 100 obserwacji")
print(pd.Series(data_100).describe())

print("Statystyki dla 10000 obserwacji")
print(pd.Series(data_10000).describe())

plt.figure(figsize=(10, 4))
sns.boxplot(data=[data_100, data_10000])
plt.xticks([0, 1], ['n=100', 'n=10000'])
plt.title('Boxplot dla dwóch rozkładów')
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
sns.histplot(data_100, bins=20, kde=True, ax=axes[0])
axes[0].set_title('Histogram – 100 obserwacji')
sns.histplot(data_10000, bins=20, kde=True, ax=axes[1])
axes[1].set_title('Histogram – 10000 obserwacji')
plt.show()
```

Podsumowanie
Mniejsze próbki są bardziej podatne na błędy większe rozwarstiwenie danych.
Większa ilość obsrwacji pokazuje lepsze odwzorowanie procesu Odchylenie 1.003 przy 0,09 dla mniejszej próbki
Histogramy powierdzają powyższe.
