```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# upload danych
df = pd.read_csv('/content/pliki/titanic.csv')

# Podstawowe statystyki opisowe
print("Statystyki zmiennej Fare:")
print(df["Fare"].describe())

plt.figure(figsize=(10, 2))
sns.boxplot(x=df["Fare"])
plt.title("wykres pudełkowy zmiennej opłata")
plt.xlabel("Cena")
plt.show()

# Histogram z estymacją gęstości
plt.figure(figsize=(8, 4))
sns.histplot(df["Fare"], bins=100, kde=True)
plt.title("Histogram cen biletów")
plt.xlabel("Opłata")
plt.ylabel("Liczba pasażerów")
plt.show()

# komentarze
print("Wnioski:")
print("- Sprzedano 891 biletów a średnia cena biletu wynosi około 32.20 a najwyższa opłata to ponad 512.")
print("- Rozkłąd jest prawostronnie skośny a znaczna większość pasażerów ok.75% zapłaciłą mniej niż 30")
```
